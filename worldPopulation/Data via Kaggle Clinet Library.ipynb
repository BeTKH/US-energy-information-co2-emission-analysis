# Data via Kaggle Client Library

```python
from kaggle.api.kaggle_api_extended import KaggleApi
import pandas as pd
import os

print("Imports complete")


# Authentication defaults to use the config file in the predefined location.
api = KaggleApi()
api.authenticate()
```

```python
api.datasets_list()



# directories
dataset = 'iamsouravbanerjee/world-population-dataset'
out_path = 'Data/WorldPopulation'


#Download file 'world_population.csv' from the dataset
api.dataset_download_file(dataset, 'world_population.csv', out_path)
```

```python
# read into pandas df
wp_df = pd.read_csv(os.path.join(out_path, 'world_population.csv'))
wp_df
```

### Population increase 

Percent Increase =  ( Population2022 − Population2000 ) / Population2000

```python
# percentage increase from 2000 to 2022 - decimal

wp_df['PercentIncrease_2000_2022'] = round(
    (wp_df['2022 Population'] - wp_df['2000 Population']) / wp_df['2000 Population'], 3)



# sort in descending order
wp_sorted = wp_df.sort_values(by='PercentIncrease_2000_2022', ascending=False)




# Result
wp_sorted[['Country/Territory','Continent','2000 Population', '2022 Population', 'PercentIncrease_2000_2022']].head(10)
```

```python
wp_sorted[['Country/Territory','Continent','2000 Population', '2022 Population', 'PercentIncrease_2000_2022']].tail(10)
```

### Save to csv file

```python
#utput path
output_path = 'Data/WorldPopulation/PopulationWithPercentIncrease.csv'

# Save 
wp_sorted.to_csv(output_path, index=False)

print(f'DataFrame saved to {output_path}')
```
